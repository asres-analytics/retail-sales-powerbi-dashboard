## Converting transaction_date to Datetime Format
The `transaction_date` column is currently stored as an `object` (string), which limits its use in `time‑based analysis`. To enable monthly and yearly trend analysis, the column is converted to a proper datetime format using pandas.

```python
# Convert transaction_date to datetime
df_cleaned['transaction_date'] = pd.to_datetime(
    df_cleaned['transaction_date'], 
    errors='coerce'
)
```

```python
# looking the transaction_date column after conversion data type.
# looking transaction_date column after conversion

df['transaction_date'].info()
print("\n")
df_cleaned['transaction_date'].info()
```


### Comparing transaction_date Data Types Before and After Conversion

The **transaction_date column** was `object (text`) in the original dataset, meaning it was not recognized as a real date and could not support time analysis. After conversion in the cleaned dataset, it became `datetime64[ns]`, confirming successful conversion. Now it supports
 - sorting, 
 - filtering, and 
 - time-based analysis

```python
# date conversion is successful, now we can create time-based columns for analysis
print(df_cleaned['transaction_date'].head())
```

```python
# Create time-based columns for analysis

# Create time-based features for analysis
df_cleaned['year'] = df_cleaned['transaction_date'].dt.year
df_cleaned['month'] = df_cleaned['transaction_date'].dt.month
df_cleaned['year_month'] = df_cleaned['transaction_date'].dt.to_period('M').astype(str)
```

```python
# looking columns of the dataset after creating time-based columns
df_cleaned.head()
```

```python
# looking time-based columns
df_cleaned[['transaction_date', 'year', 'month', 'year_month']].head()
```
